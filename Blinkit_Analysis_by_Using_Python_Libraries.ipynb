#### **DATA CLEANING**

```python
d={'Item Fat Content':'Item_fat_content', 'Item Identifier':'Item_identifier', 'Item Type':'Item_type',
       'Outlet Establishment Year':'Outlet_establishment_yr', 'Outlet Identifier':'outlet_identifier',
       'Outlet Location Type':'Outlet_loc_type', 'Outlet Size':'Outlet_size', 'Outlet Type':'Outlet_type', 'Item Visibility':'Item_visibility',
       'Item Weight':'Item_weight'}
df.rename(columns=d,inplace=True)
df
```

```python
df['Item_fat_content'].unique()
```

```python
d={'Regular':'Regular','Low Fat':'Low Fat','low fat':'Low Fat', 'LF':'Low Fat', 'reg':'Regular'}
df['Item_fat_content']=df['Item_fat_content'].map(d)
df
```

```python
a=df['Item_weight'].value_counts().mean()
df['Item_weight']=df['Item_weight'].fillna(a)
df
```

```python
df.info()
```

## **BUSINESS REQURIMENTS**

### **KPI'S REQURIMENTS**





```python
#Total Sales
total_sales=df['Sales'].sum()

#Average Sales
avg_sales=df['Sales'].mean()

#No Of Items Sold
no_of_items_sold=df['Sales'].count()

#Average Rating
avg_rating=df['Rating'].mean()

#Display
print(f"Total Sales: ${total_sales:,.1f}")
print(f"Average Sales: ${avg_sales:,.1f}")
print(f"No Of Items Sold: {no_of_items_sold:,.1f}")
print(f"Average Rating: {avg_rating:,.1f}")
```

## **VISUALIZATION**

```python
plt.figure(figsize=(20,20),facecolor='lightgreen')
plt.gcf().suptitle('Sales Data',fontsize=40,color='red',fontstyle='italic')

plt.subplot(2,3,1)
plt.title('Total Sales by fat Content',fontsize=15)
sales_by_fat_content=df.groupby('Item_fat_content')['Sales'].sum()
plt.pie(sales_by_fat_content,labels=sales_by_fat_content.index,autopct='%1.1f%%',colors=['pink','cyan'],shadow=True,explode=[0.1,0],textprops={'fontsize':20,'fontstyle':'italic'})


plt.subplot(2,3,2)
plt.title('Total Sales by item type',fontsize=15)
Sales_by_type=df.groupby('Item_type')['Sales'].sum().sort_values(ascending=False)
a=sns.barplot(y=Sales_by_type.index,x=Sales_by_type.values,hue=Sales_by_type.values,palette='Set2')
plt.gca().set_facecolor('teal')
plt.xticks(rotation=45)
plt.ylabel('Item Type')
plt.xlabel('Total Sales')
for i in a.containers:
  a.bar_label(i)

plt.subplot(2,3,3)
plt.title('Outlet Tier by Item Fat Content',fontsize=15)
a=sns.countplot(x='Outlet_loc_type',hue='Item_fat_content',data=df,palette='Set1')
plt.gca().set_facecolor('orange')
for i in a.containers:
  a.bar_label(i)
plt.xticks(rotation=45)
plt.ylabel('Total Sales')
plt.xlabel('Outlet Tier')

plt.subplot(2,3,4)
plt.title('Total Sales by Outlst Establishment Year',fontsize=15)
sales_by_year=df.groupby('Outlet_establishment_yr')['Sales'].sum()
a=sns.barplot(x=sales_by_year.index,y=sales_by_year.values,palette='rainbow',hue=sales_by_year.index)
plt.gca().set_facecolor('violet')
plt.xticks(rotation=45)
plt.ylabel('Total Sales')
plt.xlabel('Outlet Establishment Year')
for i in a.containers:
  a.bar_label(i)

plt.subplot(2,3,5)
plt.title('Total Sales by Outlet Size',fontsize=15)
sales_by_size=df.groupby('Outlet_size')['Sales'].sum()
plt.pie(sales_by_size,labels=sales_by_size.index,autopct='%0.1f%%',colors=['red','yellow','green'],textprops={'fontsize':20,'fontstyle':'italic'},radius=1,pctdistance=0.8)
plt.pie([1],radius=0.6,colors=['lightgreen'])

plt.subplot(2,3,6)
plt.title('Total Sales by Outlet Location',fontsize=15)
sales_by_type=df.groupby('Outlet_loc_type')['Sales'].sum()
a=sns.barplot(y=sales_by_type.index,x=sales_by_type.values,hue=sales_by_type.values,palette='Set1')
plt.gca().set_facecolor('cyan')
plt.xticks(rotation=45)
plt.ylabel('Outlet Location')
plt.xlabel('Total Sales')
for i in a.containers:
  a.bar_label(i)


plt.tight_layout()
plt.show()
```
